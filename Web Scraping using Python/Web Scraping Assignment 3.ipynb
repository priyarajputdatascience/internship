#Question no.4:-Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV

```python
#Importing the required libraries:-

import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
#Creating a webdriver object:-

driver=webdriver.Chrome()
```

```python
#searching for a smartphone on flipkart.com:-

smartphone= input("Enter the searching name: ")
driver.get("https://www.flipkart.com/search?q="+smartphone)
```

```python
#Creating empty lists to store the scraped data:-

brand_name=[]
smartphone_name=[]
colour=[]
ram=[]
storage=[]
primary_camera=[]
secondary_camera=[]
display_size=[]
battery_capacity=[]
price=[]
product_url=[]
```

```python
#Scraping the details for all the search results on the first page:-

products=driver.find_elements(By.CLASS_NAME,'_1AtVbE')
```

```python
#Looping through each product:-

for product in products:
    
    #Getting the product URL:-
    try:
        link=product.find_element(By.XPATH,'.//a[@class="_1fQZEK"]')
        product_url.append(link.get_attribute("href"))
    except NoSuchElementException:
        product_url.append("-")
    
    
    #Getting the brand name:-
    try:
        brand=product.find_element(By.XPATH,".//div[@class='_4rR01T']").text.split()[0]
    except NoSuchElementException:
        brand="-"
    brand_name.append(brand)
    
    
    
    #Getting the smartphone name:-
    try:
        smartphone=" ".join(product.find_element(By.XPATH,".//div[@class='_4rR01T']").text.split()[1:])
    except NoSuchElementException:
        smartphone="-"
    smartphone_name.append(smartphone)
    
    
    #Getting the colour name:-
    try:
        colr=product.find_element(By.XPATH,"div/div/div/a/div[2]/div[1]/div[1]").text
    except NoSuchElementException:
        colr="-"
    colour.append(colr)
    
    
    #Getting the RAM and ROM element:-
    try:
        RAM_ROM=product.find_element(By.XPATH,"div/div/div/a/div[2]/div[1]/div[3]/ul/li[1]").text.split("|")
        RAM=RAM_ROM[0].strip()
        ROM=RAM_ROM[-1].strip()
    except NoSuchElementException:
        RAM="-"
        ROM="-"
    ram.append(RAM)
    storage.append(ROM)
    
    
    
    
    #Getting the primary and secondary camera:-
    try:
        primary_secondary=product.find_element(By.XPATH,"div/div/div/a/div[2]/div[1]/div[3]/ul/li[3]").text.split("|")
        primary=primary_secondary[0].strip()
        secondary=primary_secondary[-1].strip()
        
    except NoSuchElementException:
        primary="-"
        secondary="-"
    primary_camera.append(primary)
    secondary_camera.append(secondary)
    
    
    
    #Getting the display size:-
    try:
        display=product.find_element(By.XPATH,"div/div/div/a/div[2]/div[1]/div[3]/ul/li[2]").text
    except NoSuchElementException:
        display="-"
    display_size.append(display)
    
    
    #Getting the battery capacity:-
    try:
        battery=product.find_element(By.XPATH,"div/div/div/a/div[2]/div[1]/div[3]/ul/li[4]").text
    except NoSuchElementException:
        battery="-"
    battery_capacity.append(battery)
    
    
    #Getting the price:-
    try:
        pri=product.find_element(By.CLASS_NAME,'_30jeq3').text
    except NoSuchElementException:
        pri="-"
    price.append(pri)
```

```python
#Creating a data frame from the scraped data:-

df=pd.DataFrame({'Brand Name':brand_name,'Smartphone Name':smartphone_name,'Colour':colour,'RAM':ram,'Storage(ROM)':storage,'Primary Camera':primary_camera,'Secondary Camera':secondary_camera,'Display Size':display_size,'Battery Capacity':battery_capacity,'Price':price,'Product URL':product_url})


#Saving the dataframe as a CSV file:-

df.to_csv("flipkart_smartphones.csv",index=False)

df
```

Question no. 5:-Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
#Importing the required libraries:-

import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
import pandas as pd
import re
```

```python
#Creating a webdriver object:-

driver=webdriver.Chrome()
```

```python
#Go to webpage:-

driver.get("https://www.google.co.in/maps")
```

```python
#Find the search box element:-

search_box=driver.find_element(By.ID,"searchboxinput")
```

```python
#Enter the city name and press enter:-

city="Nashik"
search_box.send_keys(city)
search_box.send_keys(Keys.RETURN)
```

```python
#wait for page to load:-

driver.implicitly_wait(10)
```

```python
#Get the current URL:-

url= driver.current_url
```

```python
#Extract the latitude and longitude from the URL using regular expressions:-

pattern= r"@(-?\d+\.\d+),(-?\d+\.\d+)"
match= re.search(pattern,url)
```

```python
#If a match is found, print the coordinates:-

if match:
    latitude=match.group(1)
    longitude=match.group(2)
    print(f"The geospatial coordinates of {city} are: ")
    print(f"Latitude: {latitude}")
    print(f"Longitude: {longitude}")
else:
    print("Could not find the coordinates")
```

```python
#creating empty lists:-
comments=[]
comment_time=[]
Time=[]
No_of_likes=[]
```

```python
#scrape comments:-
cm=driver.find_elements(By.ID,'content-text')
for i in cm:
    if i.text is None:
        comments.append("--")
    else:
        comments.append(i.text)
time.sleep(3)
```

```python
#scrape time when comment was posted:-
tm=driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
for i in tm:
    Time.append(i.text)
    
for i in range(0,len(Time),2):
    comment_time.append(Time[i])
time.sleep(3)
```

```python
#scrape the comment likes:-
like=driver.find_elements(By.ID,"vote-count-middle")
for i in like:
    if i.text is None:
        No_of_likes.append("--")
    else:
        No_of_likes.append(i.text)
time.sleep(3)
```

```python
#Creating empty list and find required data:-
hostel_name=[]
distance=[]
pvt_prices=[]
dorms_price=[]
rating=[]
reviews=[]
over_all=[]
facilities=[]
description=[]
```

```python
#scraping the required information:-
for i in driver.find_elements(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div/div[4]"):

    
    try:
        name_tag=driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
        for i in name_tag:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append("-")
        
    try:
        distance_tag=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
        for i in distance_tag:
            distance.append(i.text)
    except NoSuchElementException:
        distance.append("-")
        
        
    try:
        dorms_tag=driver.find_elements(By.XPATH,'//div[2][@class="property-accommodation-price"]//strong[1][@class="current"]')
        for i in dorms_tag:
            dorms_price.append(i.text)
    except NoSuchElementException:
        dorms_price.append("-")
        
    try:
        rating_tag=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]//span[@class="number"]')
        for i in rating_tag:
            rating.append(i.text)
    except NoSuchElementException:
        rating.append("-")
        
    try:
        reviews_tag=driver.find_elements(By.XPATH,'//div[@class="review"]')
        for i in reviews_tag:
            reviews.append(i.text)
    except NoSuchElementException:
        reviews.append("-")
        
    try:
        over_tag=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]//span[@class="keyword"]')
        for i in over_tag:
            over_all.append(i.text)
    except NoSuchElementException:
        over_all.append("-") 
```

```python
#scrape all product urls:-
product_urls=[]
#scraping urls:-
url=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in url:
    product_urls.append(i.get_attribute("href"))
```

```python
len(product_urls)
```

```python
#scraping remaining information:-

for url in product_urls:
    driver.get(url)
    time.sleep(2)
        
    try:
        facil_tag=driver.find_element(By.XPATH,'//ul[@class="facility-sections"]/li[1]')
        facilities.append(facil_tag.text)
    except NoSuchElementException:
        facilities.append("-")
        
    
    try:
        des_tag=driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        description.append(des_tag.text)
    except NoSuchElementException:
        description.append("-")   
        
```
